```python
from sklearn.datasets import load_iris
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

iris = load_iris()
X, y = iris.data, iris.target

params = {
    "n_estimators": [10, 50, 100],
    "max_depth": [None, 5, 10]
}

grid = GridSearchCV(RandomForestClassifier(random_state=42), params, cv=5)
grid.fit(X, y)

print(grid.best_params_)
```

Iris is a classic toy dataset used to learn and test machine learning models.

It contains 150 flower samples from 3 iris species:
Setosa
Versicolor
Virginica
Each sample has 4 numeric features:
Sepal length
Sepal width
Petal length
Petal width
In your notebook, load_iris() gives:

X: the 4-feature measurements (input data)
y: the species label (target/class) for each row
So your GridSearch is training a classifier to predict flower species from those measurements.
